```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```python
from math import exp
alpha = 0.5

elu = lambda x, alpha: x if x>0 else alpha*(np.exp(x)-1) 
y = [elu(x_i, alpha) for x_i in x]
plot_graph(y, "ELU(X)")
```

```python
alpha = 1.67326
scale = 1.0507

selu = lambda x, alpha, scale: scale*(x if x>0 else alpha*(np.exp(x)-1))
y = [selu(x_i, alpha, scale) for x_i in x]
plot_graph(y, "SELU(X)")
```
